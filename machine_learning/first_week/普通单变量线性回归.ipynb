```python
#实用单变量线性回归进行预测
import numpy as np
import matplotlib.pyplot as plt
```

```python
#生成1000个单变量数据集x
test = np.random.rand(1,1000)
one = np.ones((1,1000))
x = np.insert(test,0,values=one,axis=0)
```

```python
#生成单变量theta
theta = np.random.rand(2,1)
```

```python
#生成标签y
y = 3*test+np.random.rand(1,1000)
```

```python
#计算损失函数
def loss(x,y,theta):
    z = np.dot(theta.T,x)
    cost = (1/(2*np.shape(x)[1]))*(np.sum(np.dot(z-y,(z-y).T)))
    return cost
```

```python
#单步梯度下降
def gradient_decent(x,y,theta,learning_rate):
    z = np.dot(theta.T,x)
    theta = theta-learning_rate*(1/np.shape(x)[1])*np.dot(x,(z-y).T)
    return theta
```

```python
#进行多步迭代
losses = []
for i in range(30000):
    if(i%200==0):
        losses.append(loss(x,y,theta))
        print(losses[-1])
    theta = gradient_decent(x,y,theta,0.001)
print(theta)
```

```python
#观察数据点和拟合情况
y_predict = np.dot(theta.T,x)
plt.scatter(test,y)
plt.scatter(test,y_predict)
```

```python
#观察损失函数下降
plt.plot(losses)
```
