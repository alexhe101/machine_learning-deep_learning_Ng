```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#生成数据特征x
x = np.random.rand(1,1000)
```

```python
#生成标签y和参数
y = 4*x+np.random.rand(1,1000)
w = np.random.rand(1,1)
b = 0
```

```python
#进行前向传播
def forward(x,y,w,b):
    m  = np.shape(x)[1]
    h = np.dot(w.T,x)+b
    J = (1/(2*m))*np.sum(np.dot(h-y,(h-y).T))
    return h,J
```

```python
#进行反向传播
def backward(x,y,w,b,h,learning_rate):
    m  = np.shape(x)[1]
    dh = h-y
    dw =(1/m)*np.dot(x,dh.T)
    db = np.sum(dh)/m
    w = w -learning_rate*dw
    b = b - learning_rate*db
    return w,b
```

```python
losses= [ ]
for i in range(30000):
    h,J = forward(x,y,w,b)
    losses.append(J)
    w,b=backward(x,y,w,b,h,0.001)
    if(i%200==0):
        print(losses[-1])
```
